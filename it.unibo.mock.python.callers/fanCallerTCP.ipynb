```python
import socket

address = "localhost"
port = 8001
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

msgFanDispatch = "msg(updatefan,dispatch,python,fanactor,updatefan(STATE),X)"

def connect(port) :
    server_address = (address, port)
    sock.connect(server_address)
    print("CONNECTED ", server_address)

def forward(message, count) :
    message = message.replace("X", str(count))
    print("forward ", message)
    msg = message + "\n"
    byt = msg.encode()
    sock.send(byt)

def terminate() :
    sock.close()
    print("DISCONNECTED")

def console() : 
    count = 0
    v = str(input())
    print("INPUT", v)
    while(v != "q") :
        count += 1
        if(v == "on") :
            forward(msgFanDispatch.replace("STATE", "ON"), count)
        if(v == "off") :
            forward(msgFanDispatch.replace("STATE", "OFF"), count)
        v = str(input())
    terminate()

connect(port)
console()
```
